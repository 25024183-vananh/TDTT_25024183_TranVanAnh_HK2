#### Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a, b, c = map(float, input().split())
if (a+b>c) and (a+c>b) and (b+c>a):
    if a == b == c:
        print("Tam giac deu")
    elif (a==b) or (a==c) or (b==c):
        print("Tam giac can")
    elif (a*a + b*b == c*c) or (a*a + c*c == b*b) or (b*b + c*c == a*a):
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")
else:
    print("Khong la ba canh cua mot tam giac")
chu_vi = a+b+c
p = chu_vi/2
s = (p*(p-a)*(p-b)*(p-c))**0.5
print(f"Chu vi: {chu_vi:.2f}, Dien tich: {s:.2f}")
```

#### Bài 8: Giải hệ phương trình tuyến tính:
ax + by = m
cx + dy = n

```python
# cramer
def giai_he(a, b, m, c, d, n):
    D = a*d -b*c
    Dx = m*d - n*c
    Dy = a*n - m*c
    if D != 0:
        x = Dx/D
        y = Dy/D
        return f"Hệ phương trình có nghiệm duy nhất: x = {x}, y = {y}"
    else:
        if Dx == 0 and Dy == 0:
            return "Hệ phương trình có vô số nghiệm"
        else:
            return "Hệ phương trình vô nghiệm"
```

#### Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x0, y0 = map(float, input().split())
r = float(input())
x, y = map(float, input().split())
kc_0a = (x - x0)**2 + (y - y0)**2
if kc_0a == r**2:
    print("A thuộc đường tròn")
elif kc_0a < r**2:
    print("A nằm trong đường tròn")
else:
    print("A nằm ngoài đường tròn")
```
